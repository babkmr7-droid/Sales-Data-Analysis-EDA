```python
# Remove unnecessary columns from the dataset permanently
df.drop(['Status', 'unnamed1'],axis=1,inplace=True)
```

```python
# Remove all rows that contain missing values
df.dropna(inplace=True)

# Check again if any missing values are left
df.isnull().sum()
```

```python
# Convert Marital_Status column to string and replace 0 with Single
df['Marital_Status'] = df['Marital_Status'].astype(str).str.replace("0","Single")
```

```python
# Convert Marital_Status column to string and replace 1 with Married
df['Marital_Status'] = df['Marital_Status'].astype(str).str.replace("1","Married")
```

```python
# Check unique values available in Marital_Status column
df['Marital_Status'].unique()
```

```python
# Rename Occupation column to Sector
df.rename(columns={"Occupation":"Sector"},inplace=True)
```

```python
# Check total duplicate rows in the dataset
df.duplicated().sum()
```

```python
# Remove duplicate rows permanently from the dataset
df.drop_duplicates(inplace=True)
```

```python
# Check duplicate rows again after removing duplicates
df.duplicated().sum()
```

# ---Data Visualization---


```python
# Data Visualization is the process of representing data in graphical form.
# such as charts, graphs, and plots to understand patterns, trends, and insights easily.
# Data Visualization is a technique used to present data visually 
# using graphs, charts, and plots for better understanding and analysis.
```

```python
# -----------------Gender wise Zone Count-------------------------


# Create figure size for the chart
plt.figure(figsize=(8,6))

# Create count plot for Gender and separate bars by Zone
Ax1 = sns.countplot(x="Gender",data=df,hue="Zone",edgecolor="black")

# Add chart title with color, font weight, and font size
plt.title("Gender wise zone count", color="blue",fontweight="bold",fontsize=16)

# Remove x-axis label
plt.xlabel("")

# Remove y-axis label
plt.ylabel("")

# Add value labels on each bar
for t in Ax1.containers:
    Ax1.bar_label(t)


# Display the chart
plt.show()
```

```python
# ----------------- Gender Wise Zone Count -------------------------

# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Create figure size
plt.figure(figsize=(8,6))

# Create countplot
Ax1 = sns.countplot(
    x="Gender",
    data=df,
    hue="Zone",
    edgecolor="black",
    palette="Set2"      # Change to Set1, Set2, Set3, Dark2, tab10 etc.
)

# Add title
plt.title(
    "Gender Wise Zone Count",
    color="blue",
    fontweight="bold",
    fontsize=16
)

# Remove axis labels
plt.xlabel("")
plt.ylabel("")

# Add value labels on bars
for container in Ax1.containers:
    Ax1.bar_label(
        container,
        fmt="%d",
        fontsize=10,
        fontweight="bold"
    )

# Add legend title
plt.legend(title="Zone")

# Improve layout
plt.tight_layout()

# Show chart
plt.show()
```

```python
# --------------------Profit wise sector---------------



# Grouping data sector wise
# Here we are grouping the dataset according to Sector column
# and calculating total Orders of each sector
Top_s = df.groupby("Sector",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(15)

# Displaying top 15 sectors data
Top_s

# Creating graph size
# figsize defines width and height of graph
plt.figure(figsize=(15,5))

# Creating Line Chart
# x = Sector names
# y = Total Orders
# marker="o" adds circle points on line
# color="r" means red color line
sns.lineplot(x="Sector",y="Orders",data=Top_s,marker="o",color="r")

# Adding graph title
# color = title color
# fontweight = bold title
# fontsize = size of title
plt.title("Top 15 Sectors Profit Wise",color="blue",fontweight="bold",fontsize=16)

# Rotating x-axis labels, # So sector names can be seen clearly
plt.xticks(rotation=30)

# Adding grid lines in graph, # Makes graph easier to read
plt.grid()

# Displaying final graph
plt.show()
```

```python
# ------------------Age wise Amount--------------------


# Grouping data according to Age Group
# Here we are calculating total Amount of each age group

Pie = df.groupby("Age Group",as_index=False)["Amount"].sum()


# Creating figure size
# figsize defines width and height of chart

plt.figure(figsize=(8,6))


# Creating Pie Chart
# Pie["Amount"] = values of pie chart
# labels = names of age groups
# startangle rotates chart for better view
# autopct shows percentage on chart

plt.pie(Pie["Amount"],
        labels=Pie["Age Group"],
        startangle=42,
        autopct="%1.f%%")


# Adding chart title
# color = title color
# fontweight = bold title
# fontsize = title size

plt.title("Age Wise Amount",
          color="blue",
          fontweight="bold",
          fontsize=16)


# Displaying final chart

plt.show()
```

```python
# -------------------- State Wise Amount --------------------


# Creating graph size
# figsize defines width and height of graph

plt.figure(figsize=(15,4))


# Grouping data state wise
# Calculating total Amount of each state
# Sorting data in descending order
# head(10) selects top 10 states

S_1 = df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)


# Creating Bar Chart
# x = State names
# y = Total Amount
# data = dataset used for graph

sns.barplot(x="State",y="Amount",data=S_1)


# Adding graph title
# color = title color
# fontweight = bold title
# fontsize = title size

plt.title("Top 10 states Amount wise",
          color="blue",
          fontweight="bold",
          fontsize=16)


# Rotating x-axis labels
# Helps to display state names clearly

plt.xticks(rotation=30)


# Displaying final graph

plt.show()



# -------------------- State Wise Orders --------------------


# Creating graph size

plt.figure(figsize=(15,4))


# Grouping data according to State
# Calculating total Orders of each state
# Sorting in descending order
# Selecting top 10 states

S_2 = df.groupby("State",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)


# Creating Bar Chart
# Storing graph object into variable B_1

B_1 = sns.barplot(x="State",y="Orders",data=S_2)


# Adding graph title

plt.title("Top 10 states order wise",
          color="blue",
          fontweight="bold",
          fontsize=16)


# Rotating x-axis labels

plt.xticks(rotation=30)


# Adding labels on bars
# bar_label displays values on top of each bar

for i in B_1.containers:
    B_1.bar_label(i)


# Displaying final graph

plt.show()
```

```python
# -------------------- Marital Status Wise Amount --------------------

# Creating graph size
plt.figure(figsize=(8,5))

# Grouping data marital status wise
M_1 = df.groupby("Marital_Status",as_index=False)["Amount"].sum()

# Creating Bar Chart
sns.barplot(x="Marital_Status",y="Amount",data=M_1)

# Adding title
plt.title("Marital Status Wise Amount",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Display graph
plt.show()
```

```python
# -------------------- Product Category Wise Amount --------------------

# Creating graph size
plt.figure(figsize=(15,5))

# Grouping data according to Product Category
P_1 = df.groupby("Product_Category",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

# Creating Bar Chart
B_1 = sns.barplot(x="Product_Category",y="Amount",data=P_1)

# Adding title
plt.title("Product Category Wise Amount",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Rotating labels
plt.xticks(rotation=30)

# Display graph
plt.show()
```

```python
# -------------------- Zone Wise Orders --------------------

# Creating graph size
plt.figure(figsize=(10,5))

# Grouping data zone wise
Z_1 = df.groupby("Zone",as_index=False)["Orders"].sum()

# Creating Pie Chart
plt.pie(Z_1["Orders"],
        labels=Z_1["Zone"],
        autopct="%1.1f%%",
        startangle=90)

# Adding title
plt.title("Zone Wise Orders",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Display graph
plt.show()
```

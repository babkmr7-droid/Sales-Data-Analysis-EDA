# ---Import Libraries---

```python
# Import NumPy library for numerical operations
import numpy as np

# Import Pandas library for data handling and analysis
import pandas as pd

# Import Matplotlib for creating charts and graphs
import matplotlib.pyplot as plt

# Import Seaborn for advanced and attractive data visualization
import seaborn as sns
```

# ---Cleaned Data Load---

```python
# Load Excel dataset from the given file path into a DataFrame
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\DA_Project\Cleaned_Data_Analysis_Output.csv")
```

# ---Scatter Plot and Box Plot---


```python
# -------------------- Age and Amount Scatter Plot --------------------

# Scatter Plot is used to show the relationship between two numerical variables.
# Here we are comparing Age and Amount.
# Each point represents one customer record.
# hue="Gender" shows male and female data points in different colors.

# Creating graph size
plt.figure(figsize=(10,6))

# Creating Scatter Plot
sns.scatterplot(x="Age",
                y="Amount",
                data=df,
                hue="Gender",
                style="Gender",
                s=80)

# Adding graph title
plt.title("Age and Amount Scatter Plot",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Adding x-axis label
plt.xlabel("Age")

# Adding y-axis label
plt.ylabel("Amount")

# Display graph
plt.show()

# Explanation:
# This scatter plot helps us understand how purchase Amount changes with Age.
# If points are spread upward for any age group, it means customers of that age spent more amount.
# Gender wise color separation helps to compare spending pattern of Male and Female customers.
```

```python
# -------------------- Orders and Amount Scatter Plot --------------------

# Here we are comparing Orders and Amount.
# x = Orders placed by customer
# y = Amount spent by customer
# hue="Marital_Status" is used to compare spending pattern of Single and Married customers.

# Creating graph size
plt.figure(figsize=(10,6))

# Creating Scatter Plot
sns.scatterplot(x="Orders",
                y="Amount",
                data=df,
                hue="Marital_Status",
                style="Marital_Status",
                s=90)

# Adding graph title
plt.title("Orders and Amount Scatter Plot",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Adding x-axis label
plt.xlabel("Orders")

# Adding y-axis label
plt.ylabel("Amount")

# Display graph
plt.show()

# Explanation:
# This scatter plot shows the relationship between number of Orders and total Amount.
# It helps to check whether customers with more orders are also spending more amount.
# Marital Status wise separation helps to compare Single and Married customer behavior.
```

```python
# -------------------- Gender Wise Amount Box Plot --------------------

# Box Plot is used to understand the distribution of numerical data.
# It shows median, spread, and outliers of the data.
# Here we are checking Amount distribution according to Gender.

# Creating graph size
plt.figure(figsize=(8,6))

# Creating Box Plot
sns.boxplot(x="Gender",
            y="Amount",
            data=df)

# Adding graph title
plt.title("Gender Wise Amount Box Plot",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Adding x-axis label
plt.xlabel("Gender")

# Adding y-axis label
plt.ylabel("Amount")

# Display graph
plt.show()

# Explanation:
# This box plot helps us compare spending Amount of Male and Female customers.
# The middle line inside the box shows the median Amount.
# Points outside the box are outliers, meaning some customers spent unusually high or low amount.
```

```python
# -------------------- Age Group Wise Orders Box Plot --------------------

# Here we are checking Orders distribution according to Age Group.
# Box Plot helps to compare which age group has more variation in Orders.

# Creating graph size
plt.figure(figsize=(12,6))

# Creating Box Plot
sns.boxplot(x="Age Group",
            y="Orders",
            data=df)

# Adding graph title
plt.title("Age Group Wise Orders Box Plot",
          color="blue",
          fontweight="bold",
          fontsize=16)

# Rotating x-axis labels for better visibility
plt.xticks(rotation=30)

# Adding x-axis label
plt.xlabel("Age Group")

# Adding y-axis label
plt.ylabel("Orders")

# Display graph
plt.show()

# Explanation:
# This box plot shows how Orders are distributed in different Age Groups.
# If the box size is large, it means order variation is high in that age group.
# Outlier points show customers whose order count is very different from normal customers.
```
